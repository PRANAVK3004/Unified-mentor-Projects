```python
import pandas as pd
```

```python
df=pd.read_csv('/content/Amazon Sales data.csv')
```

```python
#displaying top 5 rows of the dataset
df.head(5)
```

```python
#checking the datatypes of all the columns
df.info()
```

```python
#here 'Order Date' and 'Ship Date' are of object type which means they are strings
#converting 'Order Date' and 'Ship Date' to datetime datatype
df['Order Date']=pd.to_datetime(df['Order Date'], format='%m/%d/%Y')
df['Ship Date']=pd.to_datetime(df['Ship Date'], format='%m/%d/%Y')
```

```python
#checking if the conversion of datatype happened or not using info()
df.info()
```

```python
#we can see that the datatype of both 'Order Date' and 'Ship Date' has been changed to datetime
#now we will export this updated file so that we can use it for analysis
df.to_csv('amazon_sales_data.csv', index=False)
```

```python
import seaborn as sns  # For data visualization
import matplotlib.pyplot as plt  # For plotting graphs
```

```python
sns.histplot(df['Total Revenue'], kde=False)
plt.title("Distribution of Total Revenue")
plt.show()
```

```python
sns.histplot(df['Total Cost'], kde=False)
plt.title("Distribution of Total Cost")
plt.show()
```

```python
sns.histplot(df['Total Profit'], kde=False)
plt.title("Distribution of Total Profit")
plt.show()
```

```python
sns.regplot(x="Total Revenue", y="Total Profit", data=df)
plt.title("Total Revenue vs Total Profit")
plt.show()
```

```python
sns.regplot(x="Units Sold", y="Total Profit", data=df)
plt.title("Units Sold vs Total Profit")
plt.show()
```

```python
sns.lmplot(x="Units Sold", y="Unit Cost", data=df, hue="Region")
plt.title("Units Sold vs Unit Cost")
plt.show()
```

```python

sns.lmplot(x="Units Sold", y="Unit Cost", data=df, hue="Sales Channel")
plt.title("Units Sold vs Unit Cost")
plt.show()
```

```python
sns.lmplot(x="Units Sold", y="Unit Cost", data=df, hue="Item Type")
plt.title("Units Sold vs Unit Cost")
plt.show()
```
